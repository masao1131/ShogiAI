```python
import torch

# GPU(CUDA)が利用可能かチェック
is_available = torch.cuda.is_available()
print(f"GPU利用可能: {is_available}")

if is_available:
    print(f"GPUデバイス名: {torch.cuda.get_device_name(0)}")
    print(f"現在のデバイス番号: {torch.cuda.current_device()}")
```

```python
# 環境のパス
print(f"作業環境：{sys.executable}")
# PyTorchのパス
print(f"PyTorchのパス：{torch.__file__}")
# PyTorchのバージョン
print(f"PyTorchのバージョン: {torch.__version__}")
```
